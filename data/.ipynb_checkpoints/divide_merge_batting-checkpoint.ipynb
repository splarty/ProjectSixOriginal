```python
import pandas as pd
batting_file = 'Batting.csv'
master = 'Master.csv'

batting_data = pd.read_csv(batting_file)
master_data = pd.read_csv(master)

grouped_hits = batting_data['H'].groupby(batting_data['playerID']).sum().reset_index()
grouped_at_bats = batting_data['AB'].groupby(batting_data['playerID']).sum().reset_index()

grouped_hits['ratio'] = grouped_hits.H.div(grouped_at_bats.AB, axis='index')

master_data = master_data.merge(grouped_hits, on='playerID', how='left')

print master_data
```
